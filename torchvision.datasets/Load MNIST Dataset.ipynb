```python
import torch
import torchvision

from torchvision import datasets
from torchvision import transforms

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
transform = transforms.Compose([transforms.ToTensor(),
                                transforms.Normalize([0.5], [1])])
```

```python
mnist_dataset = datasets.MNIST('../data', download=True, transform=transform)
mnist_dataloader = torch.utils.data.DataLoader(mnist_dataset, batch_size=64)

mnist_dataiter = iter(mnist_dataloader)
images, labels = mnist_dataiter.next()
```

```python
def imshow(img):
    img = img/2.0 + 0.5
    plt.imshow(np.transpose(img.numpy(), axes=[1, 2, 0]))
```

```python
imshow(torchvision.utils.make_grid(images))
```

```python
images, labels = mnist_dataiter.next()
imshow(torchvision.utils.make_grid(images))
```
